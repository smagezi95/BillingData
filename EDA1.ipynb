```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns


corr = df[['lifeline_units', 'total_units','above_lifeline_units']].corr()


plt.figure(figsize=(6, 4))
sns.heatmap(corr, annot=True, cmap='coolwarm', vmin=-1, vmax=1)

# Show the plot
plt.title('Correlation between lifeline_units, total_units and Above_lifeline_units')
plt.show()
```
